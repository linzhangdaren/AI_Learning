### 图像&Numpy

#### 准备工作: 

- 导入:numpy和matplotlib的pyplot模块  (缺少就用pip安装)
- 导入读取图片的模块:form PIL import Image  
- ipynb中显示图片如果出问题需要添加代码:%matplotlib inline  

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#%matplotlib inline
```

#### 代码部分

##### 打开图片

```python
#用Image打开一张图
img=Image.open('./img/color_circle.png')
```

```python
img#打开图像 在python中需要使用img.show()方法来显示图片
```

##### 转换Numpy数组

```python
#查看img类型
print(type(img))
#由于img还不是Numpy类型 需要先转化成Numpy类型
img_arr=np.asarray(img)
#array() → 强制复制（内存翻倍）
#asarray() → 智能引用（节省内存）

type(img_arr)#检查类型 类型为numpy.ndarray
print(img_arr.shape)#查看图片大小
```

##### 用plt打开Numpy格式的图片 plt.imshow(xxx)

```python
plt.imshow(img_arr)
```

#### 使用切片提取单通道

```python
#copy一份进行操作
img_arr_copy=img_arr.copy()
#查看复件是否复制成功
plt.imshow(img_arr_copy)
```

```python
#使用切片提取单通道 默认模式为色盲模式 viridis
plt.imshow(img_arr_copy[:,:,0])
```

#### 常用通道的颜色模式

颜色模列表说明:
[matplotlib官网](https://matplotlib.org/stable/users/explain/colors/colormaps.html) 

```python
#色模式为 magma
plt.imshow(img_arr_copy[:,:,0],cmap='magma')
```

在灰色模式下可以判断出是哪个通道

```python
#色模式为 gray
plt.imshow(img_arr_copy[:,:,2],cmap='gray')
```
