### OpenCV的基础使用方法

#### opencv读取图片

```python
#用cv加载一张图片 会自动转为Numpy数组
img=cv2.imread('./img/guishu.jpg')
type(img)#<class 'numpy.ndarray'>
img.shape#图片属性(高,宽,通道数量)
```

#### 显示图片 plt.imshow(img)

```python
plt.imshow(img)
```

#### OpenCv的颜色模式转换 cv2.cvtColor()

由于opencv的颜色模式默认为: BGR(读取)  
matplotlib默认的显示模式是:RGB(显示)  

导致显示结果不对所以需要转化opencv的颜色模式: BGR->RGB

```python
img_fixed=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# 2：即“to”，表示“从BGR转换到RGB”。
plt.imshow(img_fixed)
```

也可以在读取时直接选择以哪种颜色模式读取

```python
#以RGB模式读取
img_read_RGB=cv2.imread("./img/guishu.jpg",cv2.IMREAD_COLOR_RGB)
plt.imshow(img_read_RGB)
```

```python
#以灰色模式读取
img_gray=cv2.imread("./img/guishu.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
#都已经采用灰色读取了为啥还是彩色的?
#plt.imshow() 对单通道数据会自动使用彩色映射（colormap），除非显式指定灰度映射。
# plt.imshow() 中指定 cmap='gray'
plt.imshow(img_gray,cmap='gray')
```

#### 图片的缩放旋转

```python
#显示色彩模式正确的图片并查看大小
plt.imshow(img_fixed)
print(img_fixed.shape)
```

图片的缩放 cv2.resize()

```python
img_resize=cv2.resize(img_fixed, (1000,100))#(宽,高)
plt.imshow(img_resize)
```

图片的翻转 cv2.flip()

```python
img_flip=cv2.flip(img_fixed,0)
#0垂直翻转 1水平翻转 -1垂直水平翻转
plt.imshow(img_flip)
```

#### 图片的保存 cv2.imwrite(filename,img)  
保存之前要转换会CV的颜色模式

由于之前BGR转为RGB 还得再转回去RGB2BGR  
不然保存的图片颜色不正确

```python
#转换回BGR颜色模式
img_save=cv2.cvtColor(img_flip,cv2.COLOR_RGB2BGR)
```

```python
#(要保存的输出路径+图片名,要保存的图片文件名)
cv2.imwrite("./img/guishu_flip.jpg",img_save)
```
